```python
import cv2
import matplotlib.pyplot as plt

def detect_people(image_path):
    img = cv2.imread(image_path)
    #img=cv2.resize(img,(900,900))

    hog = cv2.HOGDescriptor()
    hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

    (cooredinates, w) = hog.detectMultiScale(img,
                                             winStride=(5,54),
                                             padding=(8, 8),
                                             scale=1.05)

    for (x, y, w, h) in cooredinates:
        cv2.rectangle(img, (x, y-100), (x + w, y + h), (0, 255, 0), 5)
        plt.figure(figsize=(20,20))
        plt.axis("off")
        plt.title("HOG")
        plt.imshow(img)



image_path = r"C:\Users\zbook 17 g3\Desktop\New folder (4)\Screenshot 2023-08-06 092711.png"
detect_people(image_path)
```
